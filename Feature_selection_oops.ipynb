```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

class MinMax:
    def __init__(self, data):
        self.data = data
        self.mms = MinMaxScaler()

    def data_fit(self):
        self.mms.fit(self.data[['Number of sales', 'Sales values']])

    def data_transform(self):
        transformed = self.mms.transform(self.data[['Number of sales', 'Sales values']])
        return pd.DataFrame(transformed, columns=['Number of sales', 'Sales values'])


class Standard:
    def __init__(self, data):
        self.data = data
        self.stds = StandardScaler()

    def data_fit(self):
        self.stds.fit(self.data[['Number of sales', 'Sales values']])

    def data_transform(self):
        transformed = self.stds.transform(self.data[['Number of sales', 'Sales values']])
        return pd.DataFrame(transformed, columns=['Number of sales', 'Sales values'])


class Robust:
    def __init__(self, data):
        self.data = data
        self.rbs = RobustScaler()

    def data_fit(self):
        self.rbs.fit(self.data[['Number of sales', 'Sales values']])

    def data_transform(self):
        transformed = self.rbs.transform(self.data[['Number of sales', 'Sales values']])
        return pd.DataFrame(transformed, columns=['Number of sales', 'Sales values'])
```
